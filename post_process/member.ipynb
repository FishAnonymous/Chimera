## Membership

```python
import os
import json
import json5
import pandas as pd
import re

# company = "finance_corporation"
# company = "tech_company"
company = "medical_institution"

# folder_path = '/data2/visitor/ASE25/Chimera-Dataset/Finance-Company/generated_members'
# folder_path = '/data2/visitor/ASE25/Chimera-Dataset/Tech-Company/generated_members'
folder_path = '/data2/visitor/ASE25/Chimera-Dataset/Medical-Institution/generated_members'

output_dir = f"/data2/visitor/ASE25/Chimera/Final-Output/{company}/"
if not os.path.exists(output_dir):
    os.makedirs(output_dir)
output_csv = os.path.join(output_dir, f'{company}_members.csv')

data_list = []

for filename in os.listdir(folder_path):
    if filename.endswith('.jsonc'):
        file_path = os.path.join(folder_path, filename)
        with open(file_path, 'r', encoding='utf-8') as f:
            try:
                content = json5.load(f)
                data = {
                    "name": content.get("name"),
                    "id": content.get("id"),
                    "role": content.get("role"),
                    "ip": content.get("ip"),
                    "email": content.get("email"),
                    "container_id": content.get("container_id"),
                    "mbti": content.get("mbti"),
                    "interests": content.get("interests"),
                    "personality": content.get("personality")
                }
                data_list.append(data)
            except Exception as e:
                print(f"Error loading {filename}: {e}")

# 用pandas生成DataFrame
df = pd.DataFrame(data_list)
```

```python
# save with column names
df.to_csv(output_csv, index=False, header=True)
```

# Tech Company

```python
team_json = f"/data2/visitor/ASE25/Chimera/demo/team/{company}.json"
with open(team_json, 'r', encoding='utf-8') as f:
    team_data = json.load(f)

abbr_info = {}
core_teams = team_data['core_development_team']
support_teams = team_data['support_teams']

for team_name, team_content in core_teams.items():
    # supervisor: 第一个role的abbr
    supervisor_abbr = team_content['roles'][0]['abbr']
    for role in team_content['roles']:
        abbr = role['abbr']
        abbr_info[abbr] = {
            "team_name": team_name,
            "supervisor": supervisor_abbr
        }
supervisor_abbr = support_teams['roles'][0]['abbr'] if support_teams.get('roles') else None
for role in support_teams['roles']:
    abbr = role['abbr']
    abbr_info[abbr] = {
        "team_name": "support_teams",
        "supervisor": supervisor_abbr
    }
```
