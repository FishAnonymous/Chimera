```python
import json
import os

with open("medical_institution.json", "r") as f:
    data = json.load(f)

# calculate the total number of employees
total_employees = 0

for main_team in data.values():
    for team in main_team.values():
        if isinstance(team, dict):
            for role in team["roles"]:
                total_employees += role["count"]
        else:
            # If the team is not a dictionary, it might be a list of roles
            for role in team:
                total_employees += role["count"]
print(f"Total number of employees: {total_employees}")
```
